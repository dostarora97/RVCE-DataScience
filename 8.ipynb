```python
import pandas as pd

store_data = pd.read_csv('../ds/8.csv')
```

```python
store_data.head()
```

```python
store_data.shape
```

```python
percent_display = 0
records = []
for i in range(0, 7500):
    percent_actual = ((i*100)//7499)
    if(percent_actual > percent_display) :
            percent_display = percent_actual
            print(str(percent_display)+"%")
    records.append([ str(store_data.values[i, j]) for j in range(0, 20)])
print("Completed : 100%")
```

```python
from apyori import apriori
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
print(association_results[0])
```

```python
for item in association_results:
    print("Rule: " + list(item.items)[0] + " -> " + list(item.items)[1])
    print("Support: " + str(item.support))
    print("Confidence: " + str(list(item.ordered_statistics)[0].confidence))
    print("Lift: " + str(list(item.ordered_statistics)[0].lift))
    print("=====================================")
```
